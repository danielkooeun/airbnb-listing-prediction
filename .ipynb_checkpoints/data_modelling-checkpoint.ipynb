```python
import torch
import torch.utils.data
import torch.nn as nn
from torch.nn import functional as F
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

import os
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture

GPU = torch.cuda.is_available()
```

```python
# Grab data from csv
df = pd.read_csv('aps360_airbnb2.csv', index_col=False)
df.head()
```

```python
# Cleaning Data
df_cols = list(df.columns)
df_cols.remove('price')
df_cols.remove('Unnamed: 0')
df_cols.append('price')
df = df[df_cols]

input_features = len(df_cols) - 1
print(input_features)
print(len(df))
```

```python
# turn data into 
datanp = df.values.astype(np.float32)
np.random.seed(50) # set the numpy seed for consistent split

# Data split count (train/valid/test): 2276/758/759
train_set = np.random.choice(range(3793), 2276, replace=False)
train_set.sort()
valid_set = np.random.choice([i for i in range(3793) if i not in train_set], 758, replace=False)
valid_set.sort()
test_set = [i for i in range(3793) if i not in train_set and i not in valid_set]

train_set = datanp[train_set]
valid_set = datanp[valid_set]
test_set = datanp[test_set]

separator = torch.from_numpy(np.array(range(input_features)))
```

# KMeans Stuff

```python
kmeansDistanceVecs = df[['d_airport','d_downtown','d_pikeplace','d_amazon', 'price']]
kmeansDistanceVecsTemp = (kmeansDistanceVecs ** 2)
kmeansDistanceVecs['tot'] = kmeansDistanceVecsTemp.sum(axis = 1).apply(np.sqrt)
kmeansDistanceVecs.head()
```

```python
# Group similar results together
kMeansAirport = kmeansDistanceVecs[['d_airport', 'price']]
npKMeansAirport = kMeansAirport.values
plt.scatter(npKMeansAirport[:,0],npKMeansAirport[:,1], label='True Position')  
```

```python
kMeansDowntown = kmeansDistanceVecs[['d_downtown', 'price']]
npKMeansDowntown = kMeansDowntown.values
plt.scatter(npKMeansDowntown[:,0],npKMeansDowntown[:,1])  
```

```python
kMeansPikeplace = kmeansDistanceVecs[['d_pikeplace', 'price']]
npKMeansPikeplace = kMeansPikeplace.values
plt.scatter(npKMeansPikeplace[:,0],npKMeansPikeplace[:,1])  
```

```python
kMeansAmazon = kmeansDistanceVecs[['d_amazon', 'price']]
npKMeansAmazon = kMeansAmazon.values
plt.scatter(npKMeansAmazon[:,0],npKMeansAmazon[:,1])  
```

```python
kMeanstot = kmeansDistanceVecs[['tot', 'price']]
npKMeanstot = kMeanstot.values
plt.scatter(npKMeanstot[:,0],npKMeanstot[:,1])  
```

```python
kmeans = KMeans(n_clusters=3)  
kmeans.fit(npKMeansAirport)  
plt.scatter(npKMeansAirport[:,0],npKMeansAirport[:,1], c=kmeans.labels_, cmap='rainbow')  
```

```python
kmeans2 = KMeans(n_clusters=3)  
kmeans2.fit(npKMeansDowntown)  
plt.scatter(npKMeansDowntown[:,0],npKMeansDowntown[:,1], c=kmeans2.labels_, cmap='rainbow')  
```

```python
kmeans3 = KMeans(n_clusters=3)  
kmeans3.fit(npKMeansPikeplace)  
plt.scatter(npKMeansPikeplace[:,0],npKMeansPikeplace[:,1], c=kmeans3.labels_, cmap='rainbow')  
```

```python
kmeans4 = KMeans(n_clusters=3)  
kmeans4.fit(npKMeansAmazon)  
plt.scatter(npKMeansAmazon[:,0],npKMeansAmazon[:,1], c=kmeans4.labels_, cmap='rainbow')  
```

```python
kmeans5 = KMeans(n_clusters=3)  
kmeans5.fit(npKMeanstot)  
plt.scatter(npKMeanstot[:,0],npKMeanstot[:,1], c=kmeans5.labels_, cmap='rainbow')  
```

# Gaussian Mixture Models

```python
gmm = GaussianMixture(n_components=3)
gmm.fit(npKMeansAirport)
X, Y = np.meshgrid(np.linspace(0, 1), np.linspace(0,1))
XX = np.array([X.ravel(), Y.ravel()]).T
Z = gmm.score_samples(XX)
Z = Z.reshape((50,50))
plt.contour(X, Y, Z)
plt.scatter(npKMeansAirport[:, 0], npKMeansAirport[:, 1])
plt.show()
```
